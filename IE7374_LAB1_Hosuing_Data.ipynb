```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#To compare the performances lets use sklearn.
from sklearn.linear_model import LinearRegression
from sklearn import metrics
plt.rcParams['figure.figsize'] = (12.0, 9.0)

data1 = pd.read_csv('housing_data.csv')

data1.head()
```

```python
train_df = data1.sample(frac=0.70, random_state=100)
test_df = data1[~data1.index.isin(train_df.index)]
train_df
test_df
```

```python
def z_score(df):
    # copy the dataframe
    df_std = df.copy()
    # apply the z-score method
    for column in df_std.columns:
        df_std[column] = (df_std[column] - df_std[column].mean()) / df_std[column].std()
        
    return df_std
    
# call the z_score function
df_housetrain_standardized = z_score(train_df)
df_housetest_standardized = z_score(test_df)

df_housetrain_standardized
```
